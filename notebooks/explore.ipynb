```python
import pandas as pd
df = pd.read_parquet("/home/yasida_wanigatunga/projects/nyc-taxi-batch-etl/data/raw/yellow_tripdata_2023-01.parquet")
df["dur"] = (df.tpep_dropoff_datetime - df.tpep_pickup_datetime).dt.total_seconds()/60

ms = pd.Timestamp("2023-01-01"); me = pd.Timestamp("2023-02-01")

print("out of month     :", (~df.tpep_pickup_datetime.between(ms, me, inclusive="left")).sum())
print("zero distance    :", (df.trip_distance <= 0).sum())
print("distance > 200mi :", (df.trip_distance > 200).sum())
print("negative fare    :", (df.fare_amount < 0).sum())
print("total <= 0       :", (df.total_amount <= 0).sum())
print("duration < 0.5min:", (df.dur < 0.5).sum())
print("duration > 24h   :", (df.dur > 1440).sum())
```
